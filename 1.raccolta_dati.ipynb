**CLEANING INIZIALE**

```
#filtro righe che non hanno text e username vuoto
df_tweet_completo=df_tweet_completo[(df_tweet_completo['text'].notna()) | df_tweet_completo['username'].notna()]
```

```
#elimino righe che hanno text minore di 10 caratteri
df_tweet_completo=df_tweet_completo[[len(x)>10 for x in df_tweet_completo['text']]]
```
